```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plot
%matplotlib inline
```

```python
data = pd.read_csv("data\headbrain.csv")
data = (data - data.mean())/data.std()
data.insert(0, "Bias", 1)
data.head()
```

```python
X = data.drop("Brain Weight(grams)",axis =1)
y= data["Brain Weight(grams)"]

X_train, X_valid, y_train, y_valid = train_test_split(
            X, y, test_size=0.20, random_state=42)
```

```python
class LinearRegression():
    def __init__(self, X, y, X_valid, y_valid, epochs = 1000, lr = 0.0001):
        self.X = X
        self.y = y
        self.X_valid = X_valid
        self.y_valid = y_valid
        self.epochs = epochs
        self.lr = lr        
        self.optim = GradientDesent(X, y,  epochs, lr)
        
    def fit(self):
        self.w, self.cost = self.optim.optimize()        
        y_preds = np.dot(self.X,self.w)
        
        score = self.RMSE(self.y, y_preds) 
        print(f'Model Score { score }')
        self.plotloss()
    
    def RMSE(self, y_actual, y_preds):
        return np.sqrt(np.sum(np.power((y_actual- y_preds) ,2))/len(y_actual))
    
    def plotloss(self):
        fig,ax = plot.subplots(figsize=(12,8))
        ax.plot(np.arange(self.epochs), self.cost)
        ax.set_xlabel("Epochs")
        ax.set_ylabel("Loss")
        ax.set_title("Epochs vs Loss")
        
    
    def predict(self):
        y_preds = np.dot(X_valid,self.w)
        
        score=  self.RMSE(self.y_valid.values, y_preds)
        print(f'Model Score { score }')        
        
```

```python
class GradientDesent():
    def __init__(self, X, y, epochs , lr):
        self.X =X
        self.y =y
        self.epochs = epochs
        self.lr =lr      
      
         
    def optimize(self):
        w =np.zeros(X.shape[1])          
        cost = np.zeros(self.epochs) 
        for i in range(self.epochs):
            w_grad = self.StepGrandient(w)
            w = w - (w_grad * self.lr)            
            loss = self.Loss(w) 
            cost[i] = loss            
            print (f'Epoch:{i} Loss:{loss} ')   
                
        return w, cost
                
        
    def StepGrandient(self, w):        
        m = len(self.X)            
        h = np.dot(self.X, w)
        loss = h - self.y       
        gradient = np.dot(loss, self.X)/m         
        return gradient
        
        
    def Loss(self, w):
        h = np.dot(self.X, w)
        m = len(self.X)   
        loss =  np.sum(np.power(h - self.y.values, 2))/ m   
        return loss
    
            
```

```python
model = LinearRegression(X_train, y_train,X_valid,y_valid, epochs = 1000, lr = 0.1)
```
